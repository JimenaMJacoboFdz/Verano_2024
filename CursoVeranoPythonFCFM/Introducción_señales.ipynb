# Señales periódicas

Una señal periódica es una señal que se repite de manera regular a lo largo del tiempo. Es decir, existe un intervalo de tiempo, conocido como periodo (T), después del cual la señal vuelve a tomar los mismos valores. Matemáticamente, una señal x(t) es periódica si existe un periodo T>0 tal que:

x(t+T)=x(t)

para todo valor de t.

Las señales periódicas son comunes en muchas áreas de la ciencia y la ingeniería, incluyendo la electrónica, las comunicaciones y el procesamiento de señales. Ejemplos típicos de señales periódicas incluyen las ondas senoidales, las ondas cuadradas y las ondas triangulares.

Características de una señal periódica

Periodo (T): El intervalo de tiempo después del cual la señal se repite. Se mide en segundos (s).

Frecuencia (f): El número de ciclos de la señal que ocurren en un segundo. Es el inverso del periodo y se mide en Hertz (Hz). 

f= 1/T
​
Amplitud: La magnitud máxima de la señal.

Fase: El desplazamiento horizontal de la señal en el tiempo.

# Este código genera una onda senoidal y visualiza sus características principales:

Amplitud: La altura máxima de la onda.

Periodo: El tiempo que tarda en completarse un ciclo de la onda.

Fase: El desplazamiento horizontal de la onda.

Frecuencia: La cantidad de ciclos por segundo (Hz), inversa del periodo.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la onda senoidal
amplitud = 1      # Amplitud de la onda
frecuencia = 1    # Frecuencia de la onda (Hz)
fase = 0          # Fase de la onda (radianes)
periodo = 1 / frecuencia  # Periodo de la onda (s)

# Generación del eje temporal
t = np.linspace(0, 2 * periodo, 1000)  # 2 periodos para mejor visualización

# Generación de la onda senoidal
onda = amplitud * np.sin(2 * np.pi * frecuencia * t + fase)

# Crear la figura y los ejes
fig, ax = plt.subplots()

# Graficar la onda senoidal
ax.plot(t, onda, label='Onda senoidal')

# Anotar la amplitud
ax.annotate('Amplitud',
            xy=(0.25 * periodo, amplitud), xycoords='data',
            xytext=(0.25 * periodo, amplitud + 0.5),
            arrowprops=dict(facecolor='black', shrink=0.05),
            horizontalalignment='center', verticalalignment='top')

# Anotar el periodo
ax.annotate('Periodo',
            xy=(periodo, 0), xycoords='data',
            xytext=(periodo, -1.5),
            arrowprops=dict(facecolor='red', shrink=0.05),
            horizontalalignment='center', verticalalignment='top')

# Anotar la fase
ax.annotate('Fase',
            xy=(0, 0), xycoords='data',
            xytext=(-0.2, -1),
            arrowprops=dict(facecolor='blue', shrink=0.05),
            horizontalalignment='center', verticalalignment='top')

# Configuración de la gráfica
ax.set_xlabel('Tiempo (s)')
ax.set_ylabel('Amplitud')
ax.axhline(0, color='black',linewidth=0.5)
ax.axvline(0, color='black',linewidth=0.5)
ax.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
ax.legend()

# Mostrar la gráfica
plt.title('Onda Senoidal: Amplitud, Periodo, Fase y Frecuencia')
plt.show()
```

# Frecuencia de muestreo

La frecuencia de muestreo es el número de muestras por unidad de tiempo que se toman de una señal continua para producir una señal discreta, durante el proceso necesario para convertirla de analógica en digital. Es importante porque determina la fidelidad con la que se puede reconstruir la señal original a partir de sus muestras. En este ejemplo, la frecuencia de muestreo está establecida en 50 Hz.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la onda senoidal
amplitud = 1      # Amplitud de la onda
frecuencia = 1    # Frecuencia de la onda (Hz)
fase = 0          # Fase de la onda (radianes)
periodo = 1 / frecuencia  # Periodo de la onda (s)

# Parámetro de frecuencia de muestreo
frecuencia_muestreo = 50  # Frecuencia de muestreo (Hz)

# Generación del eje temporal con la frecuencia de muestreo
t = np.arange(0, 2 * periodo, 1 / frecuencia_muestreo)  # 2 periodos para mejor visualización

# Generación de la onda senoidal
onda = amplitud * np.sin(2 * np.pi * frecuencia * t + fase)

# Crear la figura y los ejes
fig, ax = plt.subplots()

# Graficar la onda senoidal
ax.plot(t, onda, label='Onda senoidal')

# Anotar la amplitud
ax.annotate('Amplitud',
            xy=(0.25 * periodo, amplitud), xycoords='data',
            xytext=(0.25 * periodo, amplitud + 0.5),
            arrowprops=dict(facecolor='black', shrink=0.05),
            horizontalalignment='center', verticalalignment='top')

# Anotar el periodo
ax.annotate('Periodo',
            xy=(periodo, 0), xycoords='data',
            xytext=(periodo, -1.5),
            arrowprops=dict(facecolor='red', shrink=0.05),
            horizontalalignment='center', verticalalignment='top')

# Anotar la fase
ax.annotate('Fase',
            xy=(0, 0), xycoords='data',
            xytext=(-0.2, -1),
            arrowprops=dict(facecolor='blue', shrink=0.05),
            horizontalalignment='center', verticalalignment='top')

# Anotar la frecuencia de muestreo
ax.annotate('Frecuencia de muestreo: {} Hz'.format(frecuencia_muestreo),
            xy=(0.5 * periodo, -1),
            xytext=(0.5 * periodo, -1.5),
            arrowprops=dict(facecolor='green', shrink=0.05),
            horizontalalignment='center', verticalalignment='top')

# Configuración de la gráfica
ax.set_xlabel('Tiempo (s)')
ax.set_ylabel('Amplitud')
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)
ax.grid(color='gray', linestyle='--', linewidth=0.5)
ax.legend()

# Mostrar la gráfica
plt.title('Onda Senoidal: Amplitud, Periodo, Fase, Frecuencia y Frecuencia de Muestreo')
plt.show()
```

El siguiente código muestra tres gráficos, cada uno con una diferente frecuencia de muestreo (10 Hz, 20 Hz y 50 Hz). Cada gráfico compara la señal continua con la señal muestreada. Esto permite observar cómo la señal muestreada se aproxima mejor a la señal continua a medida que aumenta la frecuencia de muestreo.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la onda senoidal
amplitud = 1      # Amplitud de la onda
frecuencia = 5    # Frecuencia de la onda (Hz)
fase = 0          # Fase de la onda (radianes)
periodo = 1 / frecuencia  # Periodo de la onda (s)

# Frecuencias de muestreo para comparar
frecuencia_muestreo1 = 10  # Frecuencia de muestreo (Hz)
frecuencia_muestreo2 = 20  # Frecuencia de muestreo (Hz)
frecuencia_muestreo3 = 50  # Frecuencia de muestreo (Hz)

# Generación del eje temporal
t_continua = np.linspace(0, 2 * periodo, 1000)  # Señal continua para referencia
t1 = np.arange(0, 2 * periodo, 1 / frecuencia_muestreo1)
t2 = np.arange(0, 2 * periodo, 1 / frecuencia_muestreo2)
t3 = np.arange(0, 2 * periodo, 1 / frecuencia_muestreo3)

# Generación de la onda senoidal continua y muestreada
onda_continua = amplitud * np.sin(2 * np.pi * frecuencia * t_continua + fase)
onda1 = amplitud * np.sin(2 * np.pi * frecuencia * t1 + fase)
onda2 = amplitud * np.sin(2 * np.pi * frecuencia * t2 + fase)
onda3 = amplitud * np.sin(2 * np.pi * frecuencia * t3 + fase)

# Crear la figura y los ejes
fig, axs = plt.subplots(3, 1, figsize=(10, 8))

# Graficar la señal continua y muestreada con frecuencia_muestreo1
axs[0].plot(t_continua, onda_continua, label='Onda Continua')
axs[0].stem(t1, onda1, linefmt='r-', markerfmt='ro', basefmt='r-', label='Muestreo 10 Hz')
axs[0].set_title('Frecuencia de Muestreo: 10 Hz')
axs[0].set_xlabel('Tiempo (s)')
axs[0].set_ylabel('Amplitud')
axs[0].legend()
axs[0].grid(True)

# Graficar la señal continua y muestreada con frecuencia_muestreo2
axs[1].plot(t_continua, onda_continua, label='Onda Continua')
axs[1].stem(t2, onda2, linefmt='g-', markerfmt='go', basefmt='g-', label='Muestreo 20 Hz')
axs[1].set_title('Frecuencia de Muestreo: 20 Hz')
axs[1].set_xlabel('Tiempo (s)')
axs[1].set_ylabel('Amplitud')
axs[1].legend()
axs[1].grid(True)

# Graficar la señal continua y muestreada con frecuencia_muestreo3
axs[2].plot(t_continua, onda_continua, label='Onda Continua')
axs[2].stem(t3, onda3, linefmt='b-', markerfmt='bo', basefmt='b-', label='Muestreo 50 Hz')
axs[2].set_title('Frecuencia de Muestreo: 50 Hz')
axs[2].set_xlabel('Tiempo (s)')
axs[2].set_ylabel('Amplitud')
axs[2].legend()
axs[2].grid(True)

# Ajustar el layout
plt.tight_layout()
plt.show()
```

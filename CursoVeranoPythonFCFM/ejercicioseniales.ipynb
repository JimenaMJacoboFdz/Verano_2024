```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
from scipy.fft import rfft, rfftfreq
from scipy.signal import stft

# Cargar el archivo de audio
file_path = 'ejemplo5.wav'
sr, signal = wavfile.read(file_path)



# Transformada de Fourier (FFT)
fft_spectrum = rfft(signal)
frequencies = rfftfreq(len(signal), 1 / sr)
magnitude = np.abs(fft_spectrum)

# Transformada de Fourier de Corto Plazo (STFT)
f, t, Zxx = stft(signal, sr, nperseg=1024)

# Graficar el espectro de frecuencia (FFT)
plt.figure(figsize=(12, 8))

plt.subplot(2, 1, 1)
plt.plot(frequencies[:len(frequencies)//2], magnitude[:len(magnitude)//2])
plt.title('Espectro de Frecuencia (Transformada de Fourier)')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud')

# Graficar el espectrograma (STFT)
plt.subplot(2, 1, 2)
plt.pcolormesh(t, f, np.abs(Zxx), cmap = "magma")
plt.colorbar(label = 'Amplitud', orientation = 'horizontal')
plt.xlabel( 'Tiempo (s)')
plt.ylabel( 'Frecuencias (Hz)')
plt. tight_layout()
plt. show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# Generar una seal peridica con ruido
np.random.seed(0)  # Para reproducibilidad

# Parmetros de la seal
f_s = 1000  # Frecuencia de muestreo
T = 1.0 / f_s  # Intervalo de muestreo
t = np.linspace(0.0, 1.0, f_s, endpoint=False)  # Vector de tiempo
f = 5.0  # Frecuencia de la seal
amplitud = 1.0  # Amplitud de la seal

# Seal peridica (seno)
seal_pura = amplitud * np.sin(2 * np.pi * f * t)

# Aadir ruido gaussiano a la seal
ruido = 0.5 * np.random.normal(size=t.shape)
seal_ruidosa = seal_pura + ruido

# Visualizacin de la seal
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(t, seal_pura, label='Seal Pura')
plt.plot(t, seal_ruidosa, label='Seal Ruidosa')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.title('Seal Peridica con Ruido')
plt.legend()

# Anlisis de Fourier
N = len(t)
yf = fft(seal_ruidosa)
xf = fftfreq(N, T)[:N//2]

# Visualizacin del espectro de frecuencia
plt.subplot(2, 1, 2)
plt.plot(xf, 2.0/N * np.abs(yf[0:N//2]))
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Amplitud')
plt.title('Espectro de Frecuencia de la Seal Ruidosa')
plt.grid()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# Generar una seal peridica compuesta con ruido
np.random.seed(0)  # Para reproducibilidad

# Parmetros de la seal
f_s = 1000  # Frecuencia de muestreo
T = 1.0 / f_s  # Intervalo de muestreo
t = np.linspace(0.0, 1.0, f_s, endpoint=False)  # Vector de tiempo

# Frecuencias y amplitudes de las componentes de la seal
frecuencias = [5.0, 50.0, 120.0]  # Frecuencias en Hz
amplitudes = [1.0, 0.5, 0.3]  # Amplitudes

# Generar la seal peridica compuesta
seal_pura = sum(amplitud * np.sin(2 * np.pi * f * t) for f, amplitud in zip(frecuencias, amplitudes))

# Aadir ruido gaussiano a la seal
ruido = 0.5 * np.random.normal(size=t.shape)
seal_ruidosa = seal_pura + ruido

# Visualizacin de la seal
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(t, seal_pura, label='Seal Pura')
plt.plot(t, seal_ruidosa, label='Seal Ruidosa', alpha=0.7)
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.title('Seal Peridica Compuesta con Ruido')
plt.legend()

# Anlisis de Fourier
N = len(t)
yf = fft(seal_ruidosa)
xf = fftfreq(N, T)[:N//2]

# Visualizacin del espectro de frecuencia
plt.subplot(2, 1, 2)
plt.plot(xf, 2.0/N * np.abs(yf[0:N//2]))
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Amplitud')
plt.title('Espectro de Frecuencia de la Seal Ruidosa')
plt.grid()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# Generar una seal peridica concatenada con diferentes frecuencias en cada segmento y aadir ruido
np.random.seed(0)  # Para reproducibilidad

# Parmetros de la seal
f_s = 1000  # Frecuencia de muestreo
T = 1.0 / f_s  # Intervalo de muestreo
t_total = 3.0  # Duracin total de la seal en segundos
t = np.linspace(0.0, t_total, int(t_total * f_s), endpoint=False)  # Vector de tiempo total

# Definir segmentos de la seal
segmentos = [
    (0.0, 1.0, 5.0),   # (inicio, fin, frecuencia)
    (1.0, 2.0, 50.0),
    (2.0, 3.0, 120.0)
]

# Generar la seal concatenada
seal_pura = np.zeros_like(t)
for inicio, fin, frecuencia in segmentos:
    t_segmento = np.linspace(inicio, fin, int((fin - inicio) * f_s), endpoint=False)
    seal_pura[int(inicio * f_s):int(fin * f_s)] = np.sin(2 * np.pi * frecuencia * t_segmento)

# Aadir ruido gaussiano a la seal
ruido = 0.5 * np.random.normal(size=t.shape)
seal_ruidosa = seal_pura + ruido

# Visualizacin de la seal
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(t, seal_pura, label='Seal Pura')
plt.plot(t, seal_ruidosa, label='Seal Ruidosa', alpha=0.7)
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.title('Seal Concatenada con Diferentes Frecuencias y Ruido')
plt.legend()

# Anlisis de Fourier
N = len(t)
yf = fft(seal_ruidosa)
xf = fftfreq(N, T)[:N//2]

# Visualizacin del espectro de frecuencia
plt.subplot(2, 1, 2)
plt.plot(xf, 2.0/N * np.abs(yf[0:N//2]))
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Amplitud')
plt.title('Espectro de Frecuencia de la Seal Ruidosa')
plt.grid()

plt.tight_layout()
plt.show()
```

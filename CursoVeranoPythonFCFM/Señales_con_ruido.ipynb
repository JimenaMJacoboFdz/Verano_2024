```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# Generar una señal periódica con ruido
np.random.seed(0)  # Para reproducibilidad

# Parámetros de la señal
f_s = 1000  # Frecuencia de muestreo
T = 1.0 / f_s  # Intervalo de muestreo
t = np.linspace(0.0, 1.0, f_s, endpoint=False)  # Vector de tiempo
f = 5.0  # Frecuencia de la señal
amplitud = 1.0  # Amplitud de la señal

# Señal periódica (seno)
señal_pura = amplitud * np.sin(2 * np.pi * f * t)

# Añadir ruido gaussiano a la señal
ruido = 0.5 * np.random.normal(size=t.shape)
señal_ruidosa = señal_pura + ruido

# Visualización de la señal
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(t, señal_pura, label='Señal Pura')
plt.plot(t, señal_ruidosa, label='Señal Ruidosa')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.title('Señal Periódica con Ruido')
plt.legend()

# Análisis de Fourier
N = len(t)
yf = fft(señal_ruidosa)
xf = fftfreq(N, T)[:N//2]

# Visualización del espectro de frecuencia
plt.subplot(2, 1, 2)
plt.plot(xf, 2.0/N * np.abs(yf[0:N//2]))
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Amplitud')
plt.title('Espectro de Frecuencia de la Señal Ruidosa')
plt.grid()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# Generar una señal periódica compuesta con ruido
np.random.seed(0)  # Para reproducibilidad

# Parámetros de la señal
f_s = 1000  # Frecuencia de muestreo
T = 1.0 / f_s  # Intervalo de muestreo
t = np.linspace(0.0, 1.0, f_s, endpoint=False)  # Vector de tiempo

# Frecuencias y amplitudes de las componentes de la señal
frecuencias = [5.0, 50.0, 120.0]  # Frecuencias en Hz
amplitudes = [1.0, 0.5, 0.3]  # Amplitudes

# Generar la señal periódica compuesta
señal_pura = sum(amplitud * np.sin(2 * np.pi * f * t) for f, amplitud in zip(frecuencias, amplitudes))

# Añadir ruido gaussiano a la señal
ruido = 0.5 * np.random.normal(size=t.shape)
señal_ruidosa = señal_pura + ruido

# Visualización de la señal
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(t, señal_pura, label='Señal Pura')
plt.plot(t, señal_ruidosa, label='Señal Ruidosa', alpha=0.7)
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.title('Señal Periódica Compuesta con Ruido')
plt.legend()

# Análisis de Fourier
N = len(t)
yf = fft(señal_ruidosa)
xf = fftfreq(N, T)[:N//2]

# Visualización del espectro de frecuencia
plt.subplot(2, 1, 2)
plt.plot(xf, 2.0/N * np.abs(yf[0:N//2]))
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Amplitud')
plt.title('Espectro de Frecuencia de la Señal Ruidosa')
plt.grid()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# Generar una señal periódica concatenada con diferentes frecuencias en cada segmento y añadir ruido
np.random.seed(0)  # Para reproducibilidad

# Parámetros de la señal
f_s = 1000  # Frecuencia de muestreo
T = 1.0 / f_s  # Intervalo de muestreo
t_total = 3.0  # Duración total de la señal en segundos
t = np.linspace(0.0, t_total, int(t_total * f_s), endpoint=False)  # Vector de tiempo total

# Definir segmentos de la señal
segmentos = [
    (0.0, 1.0, 5.0),   # (inicio, fin, frecuencia)
    (1.0, 2.0, 50.0),
    (2.0, 3.0, 120.0)
]

# Generar la señal concatenada
señal_pura = np.zeros_like(t)
for inicio, fin, frecuencia in segmentos:
    t_segmento = np.linspace(inicio, fin, int((fin - inicio) * f_s), endpoint=False)
    señal_pura[int(inicio * f_s):int(fin * f_s)] = np.sin(2 * np.pi * frecuencia * t_segmento)

# Añadir ruido gaussiano a la señal
ruido = 0.5 * np.random.normal(size=t.shape)
señal_ruidosa = señal_pura + ruido

# Visualización de la señal
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(t, señal_pura, label='Señal Pura')
plt.plot(t, señal_ruidosa, label='Señal Ruidosa', alpha=0.7)
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.title('Señal Concatenada con Diferentes Frecuencias y Ruido')
plt.legend()

# Análisis de Fourier
N = len(t)
yf = fft(señal_ruidosa)
xf = fftfreq(N, T)[:N//2]

# Visualización del espectro de frecuencia
plt.subplot(2, 1, 2)
plt.plot(xf, 2.0/N * np.abs(yf[0:N//2]))
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Amplitud')
plt.title('Espectro de Frecuencia de la Señal Ruidosa')
plt.grid()

plt.tight_layout()
plt.show()
```
